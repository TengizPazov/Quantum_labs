```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. N - Nφ от T
ax1.scatter(T, N_Nφ, color='blue', s=50, alpha=0.7)
ax1.set_xlabel('T', fontsize=12)
ax1.set_ylabel('N - Nφ', fontsize=12)
ax1.set_title('N - Nφ от T', fontsize=14)
ax1.grid(True, alpha=0.3)

# 2. N - Nφ от p
ax2.scatter(p, N_Nφ, color='red', s=50, alpha=0.7)
ax2.set_xlabel('p, k3B/C', fontsize=12)
ax2.set_ylabel('N - Nφ', fontsize=12)
ax2.set_title('N - Nφ от p', fontsize=14)
ax2.grid(True, alpha=0.3)

# 3. N - Nφ от J_A
ax3.scatter(J_A, N_Nφ, color='green', s=50, alpha=0.7)
ax3.set_xlabel('J.A', fontsize=12)
ax3.set_ylabel('N - Nφ', fontsize=12)
ax3.set_title('N - Nφ от J.A', fontsize=14)
ax3.grid(True, alpha=0.3)

# 4. mkFermi от T
ax4.scatter(T, mkFermi, color='purple', s=50, alpha=0.7, label='Экспериментальные точки')


#точки для экстраполяции
selected_T = [338.4, 388.4, 439.8, 492.4]
selected_mkFermi = [147.8605, 129.3879, 108.2263, 63.7779]
selected_indices = [T.index(t) for t in selected_T]

# линейная регрессия
slope, intercept, r_value, p_value, std_err = stats.linregress(selected_T, selected_mkFermi)

T_extended = np.linspace(0, 600, 100)
mkFermi_line = slope * T_extended + intercept
T_intersection = -intercept / slope

ax4.plot(T_extended, mkFermi_line, 'r--', linewidth=2, 
         label=f'Линейная аппроксимация: y = {slope:.4f}x + {intercept:.4f}')
ax4.scatter(selected_T, selected_mkFermi, color='red', s=80, marker='s', 
           label='Точки для аппроксимации', zorder=5)
ax4.scatter([T_intersection], [0], color='orange', s=100, marker='x', 
           label=f'Пересечение с осью T: T = {T_intersection:.2f}', zorder=6)

ax4.set_xlabel('T', fontsize=12)
ax4.set_ylabel('mkFermi', fontsize=12)
ax4.set_title('mkFermi от T с линейной экстраполяцией', fontsize=14)
ax4.grid(True, alpha=0.3)
ax4.legend()
plt.tight_layout()
#plt.show()
plt.savefig('graphs.png')
print("=" * 60)
print("ЛИНЕЙНАЯ АППРОКСИМАЦИЯ ПО 4 ТОЧКАМ:")
print(f"Точки для аппроксимации: T = {selected_T}")
print(f"Соответствующие значения mkFermi: {selected_mkFermi}")
print(f"Уравнение прямой: mkFermi = {slope:.6f} * T + {intercept:.6f}")
print(f"Коэффициент детерминации R² = {r_value**2:.6f}")
print(f"Точка пересечения с осью T: T = {T_intersection:.4f}")
print("=" * 60)

print("\nСтатистика по данным:")
print(f"N - Nφ: min={min(N_Nφ):.3f}, max={max(N_Nφ):.3f}, mean={np.mean(N_Nφ):.3f}")
print(f"T: min={min(T):.1f}, max={max(T):.1f}, mean={np.mean(T):.1f}")
print(f"p: min={min(p):.1f}, max={max(p):.1f}, mean={np.mean(p):.1f}")
print(f"J.A: min={min(J_A):.2f}, max={max(J_A):.2f}, mean={np.mean(J_A):.2f}")
print(f"mkFermi: min={min(mkFermi):.3f}, max={max(mkFermi):.3f}, mean={np.mean(mkFermi):.3f}")
```
